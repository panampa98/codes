```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv('../datasets/creditcardfraud/creditcard.csv')
df
```

```python
df.describe().round(2)
```

```python
# 1. Método del Rango Intercuartílico (IQR)
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1
lim_inf_iqr = Q1 - 1.5 * IQR
lim_sup_iqr = Q3 + 1.5 * IQR

df_iqr_filtered = df[(df['Amount'] >= lim_inf_iqr) & (df['Amount'] <= lim_sup_iqr)]

# 2. Método del Quantile (0.98)
quantile_inf = df['Amount'].quantile(0.02)  # Cuantil inferior
quantile_sup = df['Amount'].quantile(0.98)  # Cuantil superior

df_quantile_filtered = df[(df['Amount'] >= quantile_inf) &
                          (df['Amount'] <= quantile_sup)]

# Mostrar resultados
print(f'Datos originales: {len(df)}')
print(f'Sin outliers (IQR): {len(df_iqr_filtered)}')
print(f'Sin outliers (Quantile 0.98): {len(df_quantile_filtered)}')
```

```python
plt.style.use('seaborn-v0_8-dark')

# Función para graficar histogramas con límites
def plot_histogram(data, title, lim_inf=None, lim_sup=None):
    # Calcular los bins automáticamente
    bins = 20
    bin_width = (max(data) - min(data)) / bins
    min_val, max_val = min(data), max(data)

    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=bins, kde=False, color='royalblue',
                 edgecolor='royalblue', alpha=0.7)
    
    labels = []  # Lista para almacenar etiquetas de la leyenda

    if lim_inf is not None:
        lbl = f'Límite Inferior: {lim_inf:.2f}'
        plt.axvline(lim_inf, color='red', linestyle='--', label=lbl)
        labels.append(lbl)
    
    if lim_sup is not None:
        lbl = f'Límite Superior: {lim_sup:.2f}'
        plt.axvline(lim_sup, color='green', linestyle='--', label=lbl)
        labels.append(lbl)
    
    # Añadir texto indicando el ancho del bin, mínimo y máximo
    text_x = max(data) * 0.9            # Posición derecha del gráfico
    text_y = plt.ylim()[1] * 0.85       # Ubicación arriba
    
    plt.text(text_x, text_y, 
             f'Mín: {min_val:.2f}\nMáx: {max_val:.2f}\nAncho bin: {bin_width:.2f}', 
             fontsize=8, color='black', ha='right', fontweight='bold')
    
    plt.title(title, fontsize=12, fontweight='bold')
    plt.xlabel('Valor', fontsize=11)
    plt.ylabel('Frecuencia', fontsize=11)

    # Solo mostrar la leyenda si hay etiquetas definidas
    if labels:
        plt.legend()

    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

# Graficar distribución original
plot_histogram(df['Amount'], 'Distribución Original de los Datos')

# Graficar datos con el método IQR
plot_histogram(df_iqr_filtered['Amount'], 'Filtrado con IQR', lim_inf_iqr, lim_sup_iqr)

# Graficar datos con el método de cuantiles (0.02 - 0.98)
plot_histogram(df_quantile_filtered['Amount'], 'Filtrado con Quantiles (0.02 - 0.98)', quantile_inf, quantile_sup)
```
