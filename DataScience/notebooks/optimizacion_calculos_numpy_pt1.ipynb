```python
import pandas as pd
import numpy as np
import time

# Crear un DataFrame grande de ejemplo
n = 10**6  # Número de filas
df = pd.DataFrame({
    'A': np.random.randint(1, 100, n),
    'B': np.random.randint(1, 100, n),
})

df
```

```python
# Función para calcular el cuadrado de la diferencia
def calc_square_difference(row):
    return (row['A'] - row['B']) ** 2

# 1. Usando apply
start_apply = time.time()
df['Result_apply'] = df.apply(calc_square_difference, axis=1)
end_apply = time.time()

print(f"Tiempo usando apply: {end_apply - start_apply:.4f} segundos")

df
```

```python
# 2. Usando .values con NumPy
start_numpy = time.time()
A, B = df['A'].values, df['B'].values
df['Result_numpy'] = (A - B) ** 2
end_numpy = time.time()

print(f"Tiempo usando .values: {end_numpy - start_numpy:.4f} segundos")

df
```
