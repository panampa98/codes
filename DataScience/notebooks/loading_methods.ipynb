```python
import pandas as pd
import time
import os
import psutil

# Ruta de archivos
csv_file = '../datasets/autotel-shared-car-locations/2020_02_25.csv'
parquet_file = '../datasets/autotel-shared-car-locations/2020_02_25.parquet'

def get_memory_usage():
    return psutil.Process(os.getpid()).memory_info().rss / (1024 ** 2)  # MB

def get_file_size(filepath):
    size_mb = os.path.getsize(filepath) / (1024 ** 2)  # en MB
    return round(size_mb, 2)

# Tama√±o de los archivos
parquet_size = get_file_size(parquet_file)

# --- Lectura Parquet ---
start_parquet = time.time()
mem_parquet_before = get_memory_usage()
df_parquet = pd.read_parquet(parquet_file)
mem_parquet_after = get_memory_usage()
time_parquet = time.time() - start_parquet

# --- Resultados ---
print('üß± Parquet:')
print(f'üì¶ Tama√±o: {parquet_size} MB')
print(f'‚è±Ô∏è Tiempo de lectura: {time_parquet:.2f} segundos')
print(f'üíæ Uso de memoria: {(mem_parquet_after - mem_parquet_before):.2f} MB')
df_parquet
```
