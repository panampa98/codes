```python
import plotly.express as px
import pandas as pd

pd.set_option('display.max_columns', None)

df = pd.read_csv('../datasets/superstore-dataset-final/Sample - Superstore.csv',
                 encoding="ISO-8859-1")
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
print(df.shape)
df.head(3)
```

```python
# Crear histograma
# Ancho del bin previamente calculado
n_bins = 5
bin_width = (df['Sales'].max() - df['Sales'].min())/(n_bins + 1)

# Crear histograma con bins personalizados
fig = px.histogram(df, x='Sales')

# Definir inicio, fin y tamao de los bins
fig.update_traces(xbins=dict(start=df['Sales'].min(), end=df['Sales'].max(), size=bin_width))

fig.update_layout(template='plotly_dark', height=400, bargap=0.2)
fig.show()
```

```python
# Crear scatter plot
fig = px.scatter(df, x="Sales", y="Profit", trendline="ols", color="Category")
fig.update_layout(template="plotly_dark", height=400)
fig.show()
```

```python
# Crear grfico de barras
df_grouped = df.groupby("Category", as_index=False)["Sales"].sum()

fig = px.bar(df_grouped, x="Category", y="Sales", color="Category", barmode="relative")
fig.update_layout(template="plotly_dark", height=400)
fig.show()
```

```python
# Crear grfico de lneas
df_grouped = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().reset_index()
df_grouped['Order Date'] = df_grouped['Order Date'].dt.to_timestamp()
df_grouped = df_grouped.sort_values(by='Order Date', ascending=True).reset_index(drop=True)

fig = px.line(df_grouped, x='Order Date', y='Sales', markers=True)
fig.update_layout(template="plotly_dark", height=400)
fig.show()
```
