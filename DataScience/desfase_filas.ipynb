```python
import pandas as pd

data = {
    'Fecha': [
        '2025-01-11 05:03:27', '2025-01-08 07:05:51', '2025-01-11 14:05:51',
        '2024-12-17 18:47:51', '2024-12-27 22:50:51', '2025-01-15 09:39:51',
        '2025-01-07 19:14:51', '2025-01-14 18:40:51', '2024-12-23 15:08:51',
        '2024-12-28 22:38:51', '2024-12-31 17:35:51', '2024-12-21 11:04:51',
        '2024-12-21 06:08:51', '2025-01-02 01:35:51', '2025-01-11 05:04:51'
    ],
    'Id': [
        24942603, 95822412, 95822412, 24942603, 95822412, 24942603,
        24942603, 13356886, 24942603, 95822412, 13356886, 24942603,
        13356886, 95822412, 24942603
    ],
    'Saldo': [
        85.46, 181.93, 332.89, 448.45, 212.74, 136.53,
        81.3, 347.57, 397.88, 493.35, 434.89, 439.92,
        221.06, 254.03, 425.35
    ]
}

df = pd.DataFrame(data)
df['Fecha'] = pd.to_datetime(df['Fecha'])
df
```

```python
df['SaldoAnterior_NoGrupo'] = df['Saldo'].shift(+1)
df
```

```python
df = df.sort_values(by=['Id', 'Fecha'], ascending=True).reset_index(drop=True)
df
```

```python
df['SaldoAnterior'] = df.groupby(['Id'])['Saldo'].shift(+1)
df['SaldoSiguiente'] = df.groupby(['Id'])['Saldo'].shift(-1)
df
```

```python
df['VariacionTiempo'] = df['Fecha'] - df.groupby(['Id'])['Fecha'].shift(1)
df
```

```python
df['VariacionTiempo'] = df['VariacionTiempo'].dt.seconds
df
```
