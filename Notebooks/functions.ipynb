## Creación de funciones y métodos en Python

Se presenta diferentes estilos de funciones desde una función mínima hasta funciones y métodos con firmas explícitas, anotaciones de tipo, validaciones y documentación.

### 1. Función mínima: sin parámetros ni `return`

La forma más sencilla solo agrupa instrucciones. Al no usar `return`, Python devuelve `None` de manera implícita.

```python
def greet():
    print("Hello!")

result = greet()
print("Returned value:", result)
```

### 2. Función con parámetros, sin tipos ni valores por defecto

Los argumentos son obligatorios y pueden enviarse por posición o por nombre. La función todavía no devuelve un resultado.

```python
def introduce(name, city):
    print(f"{name} lives in {city}.")

introduce("Ana", "Lima")
introduce(city="Cusco", name="Luis")
```

### 3. Función con valor de retorno

`return` entrega un valor al código que llamó a la función. Esto permite almacenarlo, transformarlo o usarlo en otra expresión.

```python
def add_numbers(first_number, second_number):
    return first_number + second_number

total = add_numbers(7, 5)
print("Total:", total)
```

### 4. Parámetros con valores por defecto

Un valor por defecto hace que el argumento sea opcional al llamar a la función. Los parámetros sin valor por defecto deben colocarse antes.

```python
def create_message(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(create_message("Marta"))
print(create_message("Marta", "Good morning"))
```

### 5. Parámetro opcional con `None`

`None` sirve para representar la ausencia intencional de un valor. Es especialmente útil cuando un valor predeterminado real se calcula dentro de la función.

```python
def format_profile(name, nickname=None):
    display_name = nickname if nickname is not None else name
    return f"Profile: {display_name}"

print(format_profile("Andrea"))
print(format_profile("Andrea", nickname="Andy"))
```

### 6. Cantidad variable de argumentos posicionales con `*args`

`*args` reúne en una tupla todos los argumentos posicionales adicionales. El nombre `args` es una convención; el asterisco es lo que activa el comportamiento.

```python
def calculate_average(*scores):
    return sum(scores) / len(scores) if scores else 0

print(calculate_average(15, 18, 20))
print(calculate_average())
```

### 7. Cantidad variable de argumentos nombrados con `**kwargs`

`**kwargs` reúne los argumentos enviados por nombre en un diccionario. Es flexible, aunque una firma explícita suele ser más fácil de entender y validar.

```python
def build_user(**attributes):
    return attributes

user = build_user(name="Carlos", age=30, active=True)
print(user)
```

### 8. Parámetros exclusivamente posicionales con `/`

Los parámetros situados antes de `/` deben enviarse por posición. Esta restricción es útil cuando sus nombres son detalles internos que no forman parte de la interfaz pública.

```python
def divide(dividend, divisor, /):
    return dividend / divisor

print(divide(20, 4))
# divide(dividend=20, divisor=4)  # Produce TypeError porque exige posiciones.
```

### 9. Parámetros exclusivamente nombrados con `*`

Los parámetros después de `*` deben escribirse por nombre. Esto evita confusiones en llamadas con varios valores similares y hace la intención más visible.

```python
def register_student(name, *, course, scholarship=False):
    return {"name": name, "course": course, "scholarship": scholarship}

student = register_student("Elena", course="Python", scholarship=True)
print(student)
# register_student("Elena", "Python")  # Produce TypeError: course debe nombrarse.
```

### 10. Anotaciones de tipo y retorno

Las anotaciones documentan los tipos esperados y ayudan a editores y analizadores estáticos. Python no las valida automáticamente durante la ejecución.

```python
def calculate_discount(price: float, discount_rate: float = 0.10) -> float:
    return price * (1 - discount_rate)

final_price: float = calculate_discount(150.0, discount_rate=0.20)
print(f"Final price: {final_price:.2f}")
```

### 11. Función robusta: firma precisa, tipos, documentación y validación

Esta versión combina un parámetro solo posicional, parámetros obligatoriamente nombrados, un parámetro opcional, anotaciones, *docstring* y validaciones con errores claros.

```python
def calculate_invoice(
    subtotal: float,
    /,
    *,
    tax_rate: float,
    discount_rate: float = 0.0,
    customer_name: str | None = None,
) -> dict[str, float | str | None]:
    """Calcula el total de una factura
    después del descuento y los impuestos."""
    if subtotal < 0:
        raise ValueError("subtotal must be non-negative")
    if not 0 <= tax_rate <= 1:
        raise ValueError("tax_rate must be between 0 and 1")
    if not 0 <= discount_rate <= 1:
        raise ValueError("discount_rate must be between 0 and 1")

    discounted_subtotal = subtotal * (1 - discount_rate)
    total = discounted_subtotal * (1 + tax_rate)
    return {
        "customer_name": customer_name,
        "subtotal": round(subtotal, 2),
        "total": round(total, 2),
    }

invoice = calculate_invoice(100.0,
                            tax_rate=0.18,
                            discount_rate=0.10,
                            customer_name="Rosa")
print(invoice)
```

### 12. Método de instancia

Un método de instancia recibe `self`, que representa al objeto concreto. Puede consultar o modificar sus atributos.

```python
class BankAccount:
    def __init__(self, owner: str, balance: float = 0.0) -> None:
        self.owner = owner
        self.balance = balance

    def deposit(self, amount: float) -> float:
        if amount <= 0:
            raise ValueError("amount must be positive")
        self.balance += amount
        return self.balance

account = BankAccount("Julia", 100.0)
print("New balance:", account.deposit(50.0))
```

### 13. Método de clase con `@classmethod`

Un método de clase recibe `cls` y trabaja con la clase, no con una instancia específica. Es frecuente usarlo como constructor alternativo.

```python
class Temperature:
    def __init__(self, celsius: float) -> None:
        self.celsius = celsius

    @classmethod
    def from_fahrenheit(cls, fahrenheit: float) -> "Temperature":
        celsius = (fahrenheit - 32) * 5 / 9
        return cls(celsius)

temperature = Temperature.from_fahrenheit(77.0)
print(f"{temperature.celsius:.1f} °C")
```

### 14. Método estático con `@staticmethod`

Un método estático pertenece conceptualmente a la clase, pero no necesita acceder a `self` ni a `cls`. Funciona como una utilidad organizada dentro de la clase.

```python
class TextValidator:
    @staticmethod
    def is_valid_email(email: str) -> bool:
        return "@" in email and "." in email.split("@")[-1]

print(TextValidator.is_valid_email("user@example.com"))
print(TextValidator.is_valid_email("invalid-email"))
```
